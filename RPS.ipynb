```python
from tkinter import *
import random

# Dictionaries and vars
outcomes = {
    "rock": {"rock": 1, "paper": 0, "scissors": 2},
    "paper": {"rock": 2, "paper": 1, "scissors": 0},
    "scissors": {"rock": 0, "paper": 2, "scissors": 1}
}

comp_score = 0
player_score = 0


# Functions
def converted_outcome(number):
    if number == 1:
        return "rock"
    elif number == 2:
        return "paper"
    elif number == 3:
        return "scissors"


def outcome_handler(user_choice):
    global comp_score
    global player_score
    random_number = random.randint(1, 3)
    computer_choice = converted_outcome(random_number)
    outcome = outcomes[user_choice][computer_choice]

    player_choice_label.config(image=image_dict[user_choice])
    computer_choice_label.config(image=image_dict[computer_choice])

    if outcome == 2:
        player_score += 1
        player_score_label.config(text="Player : " + str(player_score))
        outcome_label.config(fg="blue", text="Outcome : Player Won")
    elif outcome == 0:
        comp_score += 1
        computer_score_label.config(text="Computer : " + str(comp_score))
        outcome_label.config(fg="blue", text="Outcome : Computer Won")
    else:
        outcome_label.config(fg="blue", text="Outcome : Draw")


# Main Screen
master = Tk()
master.title("Rock Paper Scissors")

# Load images
rock_img = PhotoImage(file="rock.png").subsample(3)
paper_img = PhotoImage(file="paper.png").subsample(3)
scissors_img = PhotoImage(file="scissors.png").subsample(3)

image_dict = {"rock": rock_img, "paper": paper_img, "scissors": scissors_img}

# Labels
Label(master, text="Rock, Paper, Scissors", font=("Calibri", 24)).grid(row=0, columnspan=3, pady=10)
Label(master, text="Please select an option", font=("Calibri", 16)).grid(row=1, columnspan=3)
player_score_label = Label(master, text="Player : 0", font=("Calibri", 16))
player_score_label.grid(row=2, column=0, padx=10)
computer_score_label = Label(master, text="Computer : 0", font=("Calibri", 16))
computer_score_label.grid(row=2, column=2, padx=10)
player_choice_label = Label(master)
player_choice_label.grid(row=3, column=0, padx=10)
computer_choice_label = Label(master)
computer_choice_label.grid(row=3, column=2, padx=10)
outcome_label = Label(master, font=("Calibri", 16))
outcome_label.grid(row=3, column=1)

# Empty rows
# Label(master, text="").grid(row=4)

# Buttons
Button(master, image=rock_img, width=150, height=150, command=lambda: outcome_handler("rock")).grid(row=5, column=0, padx=10, pady=10)
Button(master, image=paper_img, width=150, height=150, command=lambda: outcome_handler("paper")).grid(row=5, column=1, padx=10, pady=10)
Button(master, image=scissors_img, width=150, height=150, command=lambda: outcome_handler("scissors")).grid(row=5, column=2, padx=10, pady=10)

# Dummy Label
# Label(master).grid(row=6)

master.mainloop()
```
